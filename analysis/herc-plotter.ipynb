```python
def plotter(ax,dat,colname):
    #x=np.linspace(0,1,len(dat)) #replace this with dates
    x=dat['date'].tolist()
    y=dat[colname].tolist()
    #ax.plot(x, y, color)
    ax.plot(x, y,'-o')
    fig.canvas.draw()
    
def update(ax,dat,colname):
    #x=np.linspace(0,1,len(dat)) #replace this with dates
    x=dat['date'].tolist()
    y=dat[colname].tolist()
    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            y=dat[colname].tolist()
            line.set_ydata(y)
    ax.relim()
    ax.autoscale_view(True,True,True)
```

# Plotting

```python
fig,ax = plt.subplots(2)
ax[0].set_ylabel('salinity')
ax[1].set_ylabel('temp')
```

```python
# initial plot

colnames=['Salinity','Temperature']

index=0

for index in range(0,2):

    colname=colnames[index]   
    filename=colname+'.samples'
    logfile='log_'+filename

    log=pd.read_csv(logfile,delimiter=' ',names=['date',colname])

    plotter(ax[index],log,colname)
```

```python
# update

def group_update(colnames):

    index = 0 

    for index in range(0,2):
        colname=colnames[index]
        filename=colname+'.samples'
        logfile='log_'+filename
        print(logfile)
        log=pd.read_csv(logfile,delimiter=' ',names=['date',colname])
        update(ax[index],log, colname)
```

```python
group_update(colnames)
```

```python
while True:
    group_update()
    output='updated '
    ts = time.time()
    st = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
    print(output+st)
    time.sleep(30)
```
